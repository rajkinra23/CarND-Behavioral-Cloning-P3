```python
'''
This notebook contains code to explore the data from our driving logs. The libraries to extract this data
are located in data_lib.py
'''
```

```python
import numpy as np
import data_lib as dl
```

```python
data = dl.get_data()
```

```python
# Unpack our data.
X_train, y_train = data

# For a sanity check, let's make sure X_train and y_train are of equal length.
assert len(X_train) == len(y_train)

# Log the data set size.
print("Training set size: %s" % str(len(X_train)))
```

```python
# Let's create some visualization of the turning angles we'll be dealing with.

import matplotlib.pyplot as plt
import scipy.stats as stats

# Sort the angles so we can map them on a gaussian, and plot them.
y_train.sort()
y_train = [y for y in map(float, y_train)]
mu = np.mean(y_train)
sigma = np.std(y_train)
pdf = stats.norm.pdf(y_train, mu, sigma)
plt.plot(y_train, pdf)

print("Average acceleration: %s" % str(mu))
print("Std acceleration: %s" % str(sigma))
```

```python
# I want to visuazliate the image stitching, so it's clear what's going on here. Let's just read the 
# first image, for example.
import cv2

image = X_train[0]
print("Image dimensions: %s" % str(image.shape))
plt.figure(figsize=(21, 7))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
